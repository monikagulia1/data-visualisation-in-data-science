```python
import pandas as pd
import numpy as np
```

```python
CinemaWeeklyVisitors=pd.read_csv("https://raw.githubusercontent.com/ChrisWalshaw/DataViz/master/Data/001354290/CinemaWeeklyVisitors.csv")
```

```python
CinemaAge=pd.read_csv("https://raw.githubusercontent.com/ChrisWalshaw/DataViz/master/Data/001354290/CinemaAge.csv")
```

```python
CinemaCapacity=pd.read_csv("https://tinyurl.com/ChrisCoDV/001354290/CinemaCapacity.csv")

CinemaMarketing=pd.read_csv("https://tinyurl.com/ChrisCoDV/001354290/CinemaMarketing.csv")

CinemaOverheads=pd.read_csv("https://tinyurl.com/ChrisCoDV/001354290/CinemaOverheads.csv")

CinemaSpend=pd.read_csv("https://tinyurl.com/ChrisCoDV/001354290/CinemaSpend.csv")
```

```python
print(CinemaWeeklyVisitors.head(3)) #'transpose the matrix
print(CinemaAge.head(3))
print(CinemaCapacity.head(3))
print(CinemaMarketing.head(3))
print(CinemaOverheads.head(3))
print(CinemaSpend.head(3))
```

```python
customer_data = pd.merge(pd.merge(CinemaMarketing, CinemaSpend, on='Id'), CinemaCapacity, on='Id')
```

```python
customer_data=pd.merge(pd.merge(customer_data,CinemaOverheads,on='Id'),CinemaAge,on='Id')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6, 6))
corr = customer_data.corr()
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(220, 20, n=200), square=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.title('CORRELATION MAP')
plt.show()
```

```python
#create scatterplot
plt.scatter(customer_data['Avg spend (£)'], customer_data['Avg age (yrs)'])

#calculate equation for trendline
z = np.polyfit(customer_data['Avg spend (£)'], customer_data['Avg age (yrs)'], 1)
p = np.poly1d(z)
plt.title('RELATION BETWEEN AVERAGE SPENDING AND AVERAGE AGE ')

plt.xlabel('Avg spend (£)')
plt.ylabel('Avg age (yrs)')
#add trendline to plot
plt.plot(customer_data['Avg spend (£)'], p(customer_data['Avg spend (£)']))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.stem(customer_data['Seating capacity'],customer_data['Marketing (£000s)'])
plt.title('MARKETING AND SEATING CAPICITY')
plt.xlabel('Seating capacity')
plt.ylabel('Marketing (£000s)')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
fig, ax = plt.subplots(4, 1, figsize=(6,6))


decomposition_for_max_visitor = seasonal_decompose(CinemaWeeklyVisitors['YBS'], model='multiplicative',)
decomposition_for_max_but_fall = seasonal_decompose(CinemaWeeklyVisitors['VJV'], model='multiplicative')
decomposition_for_closed= seasonal_decompose(CinemaWeeklyVisitors['ZQL'], model='additive')
decomposition_for_min_but_rise= seasonal_decompose(CinemaWeeklyVisitors['CCX'], model='additive')

 #
#A residual is the difference between an observed value and a predicted value in a regression model. It is calculated as: Residual = Observed value – Predicted value.
#A trend is the presence of a long-term increase or decrease in the sequence. Seasonality is a variation that occurs at specific regular intervals of less than a year. Seasonality can occur on different time spans such as daily, weekly, monthly, or yearly.
#
#A residual is the difference between an observed value and a predicted value in a regression model. It is calculated as: Residual = Observed value – Predicted value.
#A trend is the presence of a long-term increase or decrease in the sequence. Seasonality is a variation that occurs at specific regular intervals of less than a year. Seasonality can occur on different time spans such as daily, weekly, monthly, or yearly.
decomposition_for_max_visitor.trend.plot(ax=ax[0],xlabel='')
ax[0].set_title('trend of                   -YBS',x=0.8,y=0.97)
decomposition_for_max_but_fall.trend.plot(ax=ax[1],xlabel='')
ax[1].set_title('  -VJV ',x=0.95,y=0.95)
decomposition_for_closed.trend.plot(ax=ax[2],xlabel='')
ax[2].set_title('  -ZQL',x=0.95,y=0.95)
decomposition_for_min_but_rise.trend.plot(ax=ax[3],xlabel='')
ax[3].set_title('  -CCX',x=0.95,y=0.95)

```
